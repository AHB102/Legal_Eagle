Converting PDF to usable text format 

Enter the verdict to be summarised

```python
verdict_doc = "Landmark_hearing.pdf"
```

```python
import fitz

with fitz.open(verdict_doc) as doc:
  text = ''
  for page in doc:
     c_text = page.get_text()
     c_text = c_text.replace('\n', ' ')
     text += c_text
```

Save in a text file to be used for the basic implementation

```python
len(text)
```

```python
with open('text_filtered_full.txt', 'w', encoding ='utf-8') as file:
  file.write(text)
file.close()

text = ''
with open('text_filtered_full.txt', 'r', encoding = 'utf-8') as file:
   text = file.read(20000)
file.close()
```

```python
text
```

Import necessary dependencies 

```python
from langchain_cerebras import ChatCerebras
from langchain import PromptTemplate
from langchain.docstore.document import Document
from langchain.chains.summarize import load_summarize_chain
```

```python
doc = [Document(page_content=text)]
doc
```

```python
len(text)
```

```python
import os
from Crediential import API_KEY
if "CEREBRAS_API_KEY" not in os.environ:
    os.environ["CEREBRAS_API_KEY"] = API_KEY
```

```python
llm = ChatCerebras(
    model="llama3.1-70b",
    # temperature=0
)
```

```python
template = ''' You are a helpful assistant that summarises the given content with whatever fully available information .
Text : {text}
'''

prompt = PromptTemplate(
    input_variable = ['text'],
    template = template
)
```

```python
print("Number of tokens: ", llm.get_num_tokens(text))
```

```python
chain = load_summarize_chain(
    llm,
    chain_type='stuff',
    prompt=prompt,
    verbose=False

)


output_summary = chain.run(doc)
```

```python
output_summary
```

```python
with open("summary_with_stuffchain.txt", 'w') as file:
    file.write(output_summary)
file.close()
```
